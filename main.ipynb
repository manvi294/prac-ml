```python
# encoding
from sklearn.preprocessing import OrdinalEncoder
ord_enc = OrdinalEncoder()
data['class_name'] = ord_enc.fit_transform(data[['class_name']])
data
```

```python
# feature scaling
def scalef(aray) :
    mini = min(aray)
    maxi = max(aray)
    for x in range(0,len(aray)) :
        aray[x] = (aray[x] - mini)/(maxi-mini)

        
scalef(data['sepal_length'])
scalef(data['sepal_width'])
scalef(data['petal_length'])
scalef(data['petal_width'])
data
```
