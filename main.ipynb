```python
# encoding
from sklearn.preprocessing import OrdinalEncoder
ord_enc = OrdinalEncoder()
data['class_name'] = ord_enc.fit_transform(data[['class_name']])
data
```

```python
# feature scaling
def scalef(aray) :
    mini = min(aray)
    maxi = max(aray)
    for x in range(0,len(aray)) :
        aray[x] = (aray[x] - mini)/(maxi-mini)

        
scalef(data['sepal_length'])
scalef(data['sepal_width'])
scalef(data['petal_length'])
scalef(data['petal_width'])
data
```

```python
# data splitting
from sklearn.model_selection import train_test_split

X = data.drop(['class_name'], axis=1)
y = data['class_name']
print(X, y)

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, shuffle=1, random_state=100)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# model selection
# supervised classification
# multiclass classification

# """
# use: 

# -----Linear Models-----
# Logistic Regression
# Support Vector Machines

# -----Non-linear Models------
# K-Nearest Neighbours
# Kernel SVM
# NaÃ¯ve Bayes
# Decision Tree Classification
# Random Forest Classification

# """
```

```python
# general model evaluation function
from sklearn.metrics import confusion_matrix, classification_report
def model_Evaluate(model):
    y_pred = model.predict(X_test)
    print(classification_report(y_test, y_pred))
    cf_matrix = confusion_matrix(y_test, y_pred)
    categories = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
    group_names = ['True Neg','False Pos', 'False Neg','True Pos']
    group_percentages = ['{0:.2%}'.format(value) for value in cf_matrix.flatten() / np.sum(cf_matrix)]
    labels = [f'{v1}n{v2}' for v1, v2 in zip(group_names,group_percentages)]
    labels = np.asarray(labels).reshape(2,2)
    sns.heatmap(cf_matrix, annot = True, cmap = 'Blues',fmt = '',
    xticklabels = categories, yticklabels = categories)
    matplotlib.pyplot.xlabel("Predicted values", fontdict = {'size':14}, labelpad = 10)
    matplotlib.pyplot.ylabel("Actual values" , fontdict = {'size':14}, labelpad = 10)
    matplotlib.pyplot.title ("Confusion Matrix", fontdict = {'size':18}, pad = 20)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression

LRmodel = LogisticRegression(max_iter=1000, random_state=100)
LRmodel.fit(X_train, y_train)
model_Evaluate(LRmodel)
y_predlr = LRmodel.predict(X_test)

```

```python
#support vector machines
from sklearn import svm

svmModel = svm.SVC(kernel='linear')
svmModel.fit(X_train, y_train)
model_Evaluate(svmModel)
y_predsvm = svmModel.predict(X_test)

from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_predsvm))
```

```python
#knn
from sklearn.neighbors import KNeighborsClassifier  
knnModel= KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2 )  
knnModel.fit(X_train, y_train)
model_Evaluate(knnModel)
y_predknn = knnModel.predict(X_test)
```
